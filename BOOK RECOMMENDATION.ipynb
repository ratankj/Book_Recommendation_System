# Build up popularity based recommendation system

```python
rating_with_name=rating.merge(books,on='ISBN')
```

```python
rating_with_name.sample()
```

```python
rating_with_name.shape
```

```python
num_rating_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num-rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg-rating'},inplace=True)
avg_rating_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popularity_df
```

```python
popular_book=popularity_df[popularity_df['num-rating']>300].sort_values('avg-rating',ascending=False).head(50)
```

```python
popular_book
```

```python
popular_book.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_book=popular_book.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num-rating','avg-rating']]
```
